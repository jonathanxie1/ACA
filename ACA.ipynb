```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
aca = pd.read_csv("ACA Membership 2014-2015 - Roster.csv")
aca.head(5)
```

```python
split = aca["Sex"].value_counts()
label = 'female','male'
explode = (0,0.1)
colors = ['lightcoral','lightskyblue'] 
plt.pie(split,labels =label,colors=colors,autopct='%1.1f%%',
        shadow=True, startangle=70)
plt.axis('equal')
plt.show()
```

```python
gradconv = {'1 Grad':'Grad','2 Grad':'Grad','1 (grad)':'Grad','?':'Super-Senior','1':'Freshman','2':'Sophmore','3':'Junior','4':'Senior','5':"Super-Senior"}
years = aca["Year"].dropna()
years = years.map(gradconv)
yeard = years.value_counts()
label = 'Freshman','Sophmore','Junior','Senior','Super-Senior','Grad'
explode = (0.1,0,0,0,0,0)
colors = ['yellowgreen', 'mediumpurple', 'lightskyblue', 'lightcoral','gold','gray'] 
plt.pie(yeard,explode=explode,labels =label,colors=colors,autopct='%1.1f%%',
        shadow=True, startangle=70)
plt.axis('equal')
plt.show()
```

```python
majors = aca["Major"].dropna()
majors[majors.str.contains("Math")] = "Math"
majors[majors.str.contains("Communication")] = "Communications"
majors[majors.str.contains("Busi")] = "Business Economics"
majors[majors.str.contains("Comp")] = "Computer Science"
allmajors = majors.value_counts()
bigmajors=allmajors.head(10)
plt.bar(range(10), bigmajors)
plt.xticks(range(10), bigmajors.index, rotation="vertical")
plt.show()
#majors[majors["Major"].contains("math")] 
#majorstats = majors.value_counts()
#len(majorstats)
#majors
```

I combined all types of Math under Math, so in reality math is actually not that popular :(

```python

majors[majors.str.contains("Psychobiology")] = "Life Science"
majors[majors.str.contains("Neuro")] = "Life Science"
majors[majors.str.contains("Psych")] = "Life Science"
majors[majors.str.contains("Bio")] = "Life Science"
majors[majors.str.contains("MIMG")] = "Life Science"
majors[majors.str.contains("MCDB")] = "Life Science"
majors[majors.str.contains("Physiological")] = "Life Science"
majors[majors.str.contains("Econ")] = "Business Economics/Economics"
majors[majors.str.contains("Engineer")] = "Engineer"
majors[majors.str.contains("Computer")] = "Engineer"
majors[majors.str.contains("CSE")] = "Engineer"
majors[majors.str.contains("Life Science")] = "Life Science"
majors[majors.str.contains("Asi")] = "Asian Studies"
majors[majors.str.contains("Micro")] = "Life Science"
majors[majors.str.contains("Phsychobiology")] = "Life Science"
allmajors = majors.value_counts()
allmajors
LSpercent = 129.0/sum(allmajors)
BEpercent = 62.0/sum(allmajors)
Engpercent = 46.0/sum(allmajors)
Mathpercent = 21.0/sum(allmajors)
big4 = LSpercent+BEpercent+Engpercent+Mathpercent
other = 1-big4
label = 'Life Science','Bus Econ/Econ','Engineer','Math','Other'
explode = (0.1,0,0,0,0)
colors = ['yellowgreen', 'mediumpurple', 'lightskyblue', 'lightcoral','gold']
majorsplit = [LSpercent,BEpercent,Engpercent,Mathpercent,other]
plt.pie(majorsplit,explode=explode,labels =label,colors=colors,autopct='%1.1f%%',
        shadow=True, startangle=70)
plt.axis('equal')
plt.show()
```

```python
home = aca["Hometown"].dropna()
home[home.str.contains("CA")] = "California"
home[home.str.contains("China")] = "Asia"
home[home.str.contains("Shanghai")] = "Asia"
home[home.str.contains("Xi")] = "Asia"
home[home.str.contains("Tai")] = "Asia"
home[home.str.contains("Hong Kong")] = "Asia"
home[home.str.contains("Beijing")] = "Asia"
home[home.str.contains("Thailand")] = "Asia"
home[home.str.contains("Korea")] = "Asia"
home[home.str.contains("Singapore")] = "Asia"
home[home.str.contains("Canada")] = "non-Asia international"
home[home.str.contains("Montreal")] = "non-Asia international"
home[home.str.contains("Canada")] = "non-Asia international"
home[home.str.contains("New Zealand")] = "non-Asia international"
home[home.str.contains("Brazil")] = "non-Asia international"
home[home.str.contains("Indonesia")] = "Asia"
home[home.str.contains(",")] = "Out of State"

total = home.value_counts()
instate = sum(total)-66-31-8
label = 'Cali','Asia','Out of State','Non-Asia International'
explode = (0.1,0,0,0)
colors = ['yellowgreen','lightskyblue','lightcoral','gold']
homesplit = [instate,66,31,8]
plt.pie(homesplit,explode =explode,labels =label,colors=colors,autopct='%1.1f%%',
        shadow=True, startangle=70)
plt.axis('equal')
plt.show()
```
